```python
# PL 2020 Season teams

pl_teams = pl_results[["Home_Team"]]

pl_teams.Home_Team.unique() # 20 teams
```

```python
# For each team, Track number of wins and losses through each game in season 
# New columns: Home_Wins, Home_Losses, Away_Wins, Away_Losses, Draws, total_points

team = 'Liverpool'
team_pl_results = pl_results.loc[(pl_results["Home_Team"]==team) | (pl_results["Away_Team"]==team)]

team_pl_results
```

```python
# Read Result column to see if W/L/D etc
# Add in new columns to split result into Home_Score and Away_Score
home_result = []
away_result = []

for x in team_pl_results.Result:
    home_result.append(int(x[:x.find('-')]))
    away_result.append(int(x[x.find('-')+1:]))

team_pl_results["Home_Score"] = home_result
team_pl_results["Away_Score"] = away_result 

team_pl_results
```

```python
team_pl_results["Home_Team"][:10]
```

```python

for i in team_pl_results["Home_Team"]:
    if i == "Liverpool":
        team_pl_results["Team_Score"] = team_pl_results[["Home_Score"]]
    else: 
        team_pl_results["Team_Score"] = team_pl_results[["Away_Score"]]
# if team_pl_results["Away_Team"] == team:
#     team_pl_results["Team_Score"] = team_pl_results[["Away_Score"]]

```

```python
# Add in team score (easier for plotting)

# check to see if given team are home team or away team for given match and find score
team_score = []
for ind, team_name in enumerate(team_pl_results["Home_Team"]):
    if team_name == team:
        team_score.append(home_result[ind])
    else:
        team_score.append(away_result[ind])

team_pl_results["Team_Score"] = team_score
    
team_pl_results

team_pl_results.iloc[:, [0, 1, 2, 5, 7, 8, 9]]
```
